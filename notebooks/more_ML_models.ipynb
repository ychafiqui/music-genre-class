```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
dataset = pd.read_csv("music_genre_dataset_30s.csv")
dataset
```

```python
dataset = dataset.sample(frac=1, random_state=0)
dataset
```

```python
X = dataset.drop(columns=["label"])
y = dataset["label"]
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X = ss.fit_transform(X)
X
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

## **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
```

```python
print(accuracy_score(y_train, lr.predict(X_train)))
print(accuracy_score(y_test, lr.predict(X_test)))
```

## **Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
print(accuracy_score(y_train, rf.predict(X_train)))
print(accuracy_score(y_test, rf.predict(X_test)))
```

## **XGBoost**

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train, y_train)
```

```python
print(accuracy_score(y_train, xgb.predict(X_train)))
print(accuracy_score(y_test, xgb.predict(X_test)))
```

## **KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
print(accuracy_score(y_train, knn.predict(X_train)))
print(accuracy_score(y_test, knn.predict(X_test)))
```

## **SVM**

```python
from sklearn.svm import SVC
svm = SVC(probability=True)
svm.fit(X_train, y_train)
```

```python
print(accuracy_score(y_train, svm.predict(X_train)))
print(accuracy_score(y_test, svm.predict(X_test)))
```

## **CatBoost**

```python
from catboost import CatBoostClassifier
cb = CatBoostClassifier()
cb.fit(X_train, y_train, silent=True)
```

```python
print(accuracy_score(y_train, cb.predict(X_train)))
print(accuracy_score(y_test, cb.predict(X_test)))
```
